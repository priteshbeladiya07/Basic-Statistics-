# For removing Outliers

```python
import numpy as np
arr = np.array([2,3,4,6,7,8,9,23,35,56,67,84,245])
arr
```

```python
Q1 = np.percentile(arr,25)
Q1
```

```python
Q3 = np.percentile(arr,75)
Q3
```

```python
IQR =  Q3 - Q1
IQR
```

```python
UF=Q3 + (IQR*1.5)
LF = Q1- (IQR*1.5)
UF
```

```python
s=[]
for i in arr:
    if i>= LF and i<=UF:
        s.append(i)
arr2 = np.array(s)
arr2
```

# Z - test

```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean - population_mean)/(population_std / np.sqrt(n))
print(z_score)
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null hypothesis")
```

# T - test

```python
import numpy as np
from scipy import stats
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
sample_std= np.std(sample,ddof = 1)
sample_mean = np.mean(sample)
n = len(sample)
population_mean = 170
```

```python
t_stats = (sample_mean - population_mean)/ (sample_std/np.sqrt(n))
print(t_stats)
```

```python
p_value = 2* (1-stats.t.cdf(abs(t_stats),df=n-1))
print(p_value)
```

```python
alpha= 0.05
if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null hypothesis")
```
