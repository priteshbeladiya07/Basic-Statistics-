# TWO SAMPLE T - TEST

```python
import numpy as np
from scipy import stats
```

```python
G_A = [85,88,90,92,87,85,89,91,86,88]
G_B = [82,84,80,83,81,79,78,85,84,83]
```

```python
t_stats , p_value = stats.ttest_ind(G_A , G_B , equal_var = False )
print(t_stats)
```

```python
alpha = 0.05
if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null hypothesis")
```

# Chi Square

```python
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df
```

```python
contingency_table = pd.crosstab(df['sex'],df['survived'])
```

```python
contingency_table
```

```python
contingency_table
```

```python
chi2,p_value,dof,expected = chi2_contingency(contingency_table)
```

```python
dof
```

```python
print(p_value)
```

```python
print(expected)
```

```python
print(chi2)
```

```python
alpha = 0.05
if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null hypothesis")
```
